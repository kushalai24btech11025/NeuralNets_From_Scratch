```python
import numpy as np
import pandas as pd

# Reproducibility
np.random.seed(42)

# -----------------------------
# 1. Generate Base Dataset (500 samples)
# -----------------------------

num_samples = 500

student_id = np.random.randint(1, 201, num_samples)   # assume 200 students
day_of_week = np.random.randint(0, 7, num_samples)    # 0 = Monday
meal_time = np.random.randint(0, 3, num_samples)      # 0 breakfast,1 lunch,2 dinner

meal_type = np.random.randint(0, 2, num_samples)      # 0 veg,1 non-veg

# Weekend calculation
is_weekend = (day_of_week >= 5).astype(int)

# Class day (0 = class, 1 = no class)
class_day = np.random.choice([0, 1], num_samples, p=[0.7, 0.3])

# Assignment deadline (rare)
assignment_deadline = np.random.choice([0, 1], num_samples, p=[0.9, 0.1])

# Weather features
temperature = np.random.normal(28, 5, num_samples)     # °C
humidity = np.random.uniform(40, 90, num_samples)      # %
wind_speed = np.random.uniform(0, 20, num_samples)     # km/h
rain = np.random.choice([0, 1], num_samples, p=[0.8, 0.2])
air_quality = np.random.uniform(50, 200, num_samples)  # AQI

# Lifestyle features
rising_time = np.random.normal(420, 60, num_samples)   # around 7:00 AM (minutes)
sleeping_time = np.random.normal(1380, 60, num_samples) # around 11:00 PM

# Target variable (mess duration)
mess_duration = (
    20
    + meal_time * 10
    + meal_type * 5
    + is_weekend * 8
    + assignment_deadline * 10
    + rain * 5
    + np.random.normal(0, 5, num_samples)
)

# Create DataFrame
data = pd.DataFrame({
    "student_ID": student_id,
    "day_of_week": day_of_week,
    "meal_time": meal_time,
    "meal_type": meal_type,
    "is_weekend": is_weekend,
    "class_day": class_day,
    "assignment_deadline": assignment_deadline,
    "temperature": temperature,
    "humidity": humidity,
    "wind_speed": wind_speed,
    "rain": rain,
    "air_quality": air_quality,
    "rising_time": rising_time,
    "sleeping_time": sleeping_time,
    "mess_duration": mess_duration
})

# -----------------------------
# 2. Extend Dataset to 5000 Samples
# -----------------------------

target_size = 5000
extra_samples = target_size - len(data)

# Sample rows with replacement
augmented = data.sample(extra_samples, replace=True).copy()

# Add Gaussian noise
noise_scale = {
    "temperature": 1.5,
    "humidity": 3,
    "wind_speed": 1,
    "air_quality": 5,
    "rising_time": 10,
    "sleeping_time": 10
}

for col, scale in noise_scale.items():
    augmented[col] += np.random.normal(0, scale, extra_samples)

# Combine datasets
final_dataset = pd.concat([data, augmented], ignore_index=True)

# -----------------------------
# 3. Save Dataset
# -----------------------------

final_dataset.to_csv("iit_h_mess_dataset.csv", index=False)

print("Dataset generated successfully!")
print("Total samples:", len(final_dataset))
```
